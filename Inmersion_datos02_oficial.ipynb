#Clase 02 tratamiento de datos y primeros graficos

```python
from google.colab import drive

drive.mount('/content/drive')

#importando pandas
import pandas as pd

#leer documento y pasar parametro de la ruta
inmuebles = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/inmersión de datos con python/inmuebles_bogota.csv')

#mostrar solo 5 lineas
inmuebles.head()
```

```python
#saber el tipo de dato
type(inmuebles.Valor[0])
```

```python
#concatenando valores del 0 al 5
inmuebles.Valor[0] + inmuebles.Valor[1] + inmuebles.Valor[:5]
```

```python
inmuebles.Valor[0].split()  #lista de dos elementos, separando la cadena de texto
```

```python
#convirtiendo a string
type(inmuebles.Valor.str.split())
```

```python
valor = inmuebles.Valor.str.split(expand=True)
#columnas moneda y precio
inmuebles['Moneda'] = valor[0] #nuevos atributos
inmuebles['Precio'] = valor[1]
inmuebles.sample(5)
```

```python
inmuebles.info()
```

```python
#concatenar
inmuebles['Precio'].sum()
```

```python
#reemplazar los puntos en el precio
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.', '', regex=True )
```

```python
inmuebles[['Precio', 'Barrio']]
```

```python
inmuebles.info()
```

```python
#disminuir la cantidad de ceros en el precio
inmuebles['Precio_millon'] = inmuebles.Precio.astype('float')/1000000
```

```python
#describir los datos
inmuebles.describe()
```

```python
#ajustar los decimales
pd.set_option('display.precision', 2)
pd.set_option('display.float_format', lambda x : '%.2f' % x)
inmuebles.describe()
```

```python
#devolviendo valores
inmuebles.loc[inmuebles.Habitaciones == 110]
```

```python
inmuebles.loc[inmuebles.Área == 2]
```

```python
#grafica histograma
inmuebles['Precio_millon'].plot.hist(bins=100)
```

```python
#importando
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))   #tamaño del lienzo
grafica = sns.histplot(data=inmuebles, x= 'Precio_millon', kde=True, hue='Tipo')
grafica.set_title('Distribucion de valores de los inmuebles en Bogota')

plt.xlim((50, 1000)) #limite de variable (x)
#plt.ylim((0, 100))  #limite de variable (y)

plt.savefig('/content/drive/MyDrive/Colab Notebooks/inmuebles_Valor.png', format='png') #guardar imagen de la grf

plt.show()
```

```python
#variable
inmuebles_cant = inmuebles[inmuebles.Tipo.isin(['Bodega', 'Local',])]
inmuebles_cant.head(10)
```

```python
plt.figure(figsize=(10,6))
grafica_desafio = sns.histplot(data=inmuebles_cant, x='Precio_millon', kde=True, hue='Tipo')

grafica_desafio.set_title('Distribucion de la cantidad de inmuebles y precios')
plt.xlim((50,6000)) #limite
plt.xlabel('Precio en Pesos Colombianos')
plt.ylabel('Cantidad de Inmuebles')

plt.show()
```
