```python
#ejecucion en grafica
inmuebles_barrio= inmuebles.Barrio.value_counts()
inmuebles_barrio.plot.bar()
```

```python
#obtenemos los 10 barrios
inmuebles_barrio = inmuebles.Barrio.value_counts()
inmuebles_barrio.head(10).plot.bar(color='pink')
```

```python
#Promedio de área de todos los inmuebles en los barrios en el dataset. El top 10.
grupo = inmuebles.groupby(['Barrio'])['Area'].mean()
grupo.head(10).plot.bar()
```

```python
#Consultar otros datos estadísticos, conteo, mediana, valores mínimo y máximo.
grupo = inmuebles.groupby(['Barrio'])['Area'].value_counts()
grupo.head(8).plot.bar()
```
